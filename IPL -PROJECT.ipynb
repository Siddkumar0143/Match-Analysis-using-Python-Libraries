```python
# Loading the libraries
import pandas as pd  # data manipulation
from matplotlib import pyplot as plt # for data visualization
import seaborn as sns
```

```python
# Import/load IPL matches dataset i.e. csv format   comma separated values
matches=pd.read_csv('matches.csv')
```

```python
# to get first 5 rows of datasets
matches.head()
```

```python
#to know no. of rows and no. of columns in the datasets
matches.shape
```

```python
# to get general information of the datasets
matches.describe()  # there are different types of data like- float,strings resent in the datasets
```

```python
# How many seasons we have got in the dataset?
matches['date'].unique()
```

```python
# The number of matches played in each session?
matches['date'].value_counts()
```

```python
#Result based on season years
matches['result'].value_counts()
```

```python
# Number of matches played in each venue
matches['venue'].value_counts()
```

```python
# Match winners based on Season year
matches.groupby('date')['winner'].value_counts()   # group-by function - split the data into groups based on some criteria. 
```

```python
#Exploring with dates,months,years


matches['date']=pd.to_datetime(matches['date'])   # from date we want to extract  the dates,months,years so we use pandas datetime to do this
```

```python
#to extract month from date

matches['month']=matches['date'].dt.month
```

```python
#to extract years
matches['year']=matches['date'].dt.year
```

```python
# No. of incidents where the team has chosen to bat or field?
matches['toss_decision'].value_counts()
```

```python
#How many Man of the matches the players are awarded amongst all the seasons?
matches['player_of_match'].value_counts()
```

```python
# Most number of match held in a city
matches['city'].value_counts()
```

```python
# Let's see some visualiisations using python
#Matplotlib library is used for data visualization.
# we can create various plots like scattered plot,bar plot,box plot,etc.
```

```python
# No. of matches played in each season of IPL(2008-2020)
plt.hist(matches['year'],bins=20)   #use histogram to plot these data
```

```python
#seaborn is a data visualization library based on matplotlib.It provides a high=level interface for drawing attractive and informative statistical graphics. 
```

```python
# Which team is most successful among all the seasons?
import seaborn as sns
sns.countplot(y='winner',data=matches)
```
